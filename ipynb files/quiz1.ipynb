```python
import math
import pandas as pd

data = pd.read_csv('iris.csv')

def linear_regre(arrx,arry,n):
    sumx = 0
    sumy = 0
    sumxy = 0
    sumx2 = 0
    sumy2 = 0
    for i in range(n):
        sumx += arrx[i]
        sumy += arry[i]
        sumxy += arrx[i] * arry[i]
        sumx2 += arrx[i]**2
        sumy2 += arry[i]**2
    dividend = (n * sumxy) - (sumx * sumy)
    divisor =  n*sumx2-sumx**2
    b = dividend/divisor
    meanx = sumx/n
    meany = sumy/n
    a = meany - b * meanx
    c_dividend = (n * sumxy) - (sumx * sumy)
    c_divisor = math.sqrt(((n* sumx2) - sumx**2) * ((n * sumy2) - sumy**2))
    r= c_dividend / c_divisor
    print(f"y = ({round(a, 4)}) - ({round(b, 4)})x")
    print(f"Linear Regression is {round(r, 4)}")
    
def exp_regre(arrx,arry,n):
    sumx = 0
    sumy = 0
    sumxy = 0
    sumx2 = 0
    sumy2 = 0
    for i in range(n):
        sumx += arrx[i]
        sumy += math.log(arry[i])
        sumxy +=  math.log(math.pow(arry[i],arrx[i]))
        sumx2 += arrx[i]**2
        sumy2 += math.log(arry[i])**2
    dividend = (n * sumxy) - (sumx * sumy)
    divisor =  n*sumx2-sumx**2
    b = dividend/divisor
    meanx = sumx/n
    meany = sumy/n
    a = meany - b * meanx
    A = pow((math.e),a) 
    dividend_cor = (n * sumxy) - (sumx * sumy)
    divisor_cor = math.sqrt(((n* sumx2) - sumx**2) * ((n * sumy2) - sumy**2))
    r= dividend_cor / divisor_cor
    print(f"y = {round(A, 4)}e^({round(b, 4)}x)",)
    print(f"Exponential Regression is {round(r, 4)}")
    
def log_regre(arrx,arry,n):  
    sumx = 0
    sumy = 0
    sumxy = 0
    sumx2 = 0
    sumy2 = 0
    for i in range(n):
        sumx += math.log10(arrx[i])
        sumy += math.log10(arry[i])
        sumxy += math.log10(arrx[i]) * math.log10(arry[i])
        sumx2 += math.log10(arrx[i])**2
        sumy2 += math.log10(arry[i])**2
    #B 
    dividend = (n * sumxy) - (sumx * sumy)
    divisor =  n*sumx2-sumx**2
    b = dividend/divisor
    #A
    meanx = sumx/n
    meany = sumy/n
    A = meany - b * meanx
    a = math.pow(10,A)
    dividend_cor = (n * sumxy) - (sumx * sumy)
    divisor_cor = math.sqrt(((n* sumx2) - sumx**2) * ((n * sumy2) - sumy**2))
    r= dividend_cor / divisor_cor
    print(f"y = ({round(a, 4)}x^({round(b, 4)})")
    print(f"Logarithmic Regression is {round(r, 4)}")
    
def hyp_regre(arrx, arry,n):
    sumx = 0
    sumy = 0
    sumxy = 0
    sumx2 = 0
    sumy2 = 0
    for i in range(n):
        sumx += 1/arrx[i]
        sumy += 1/arry[i]
        sumxy += (1/arrx[i]) * (1/arry[i])
        sumx2 += (1/arrx[i])**2
        sumy2 += (1/arry[i])**2
    dividend = (n * sumxy) - (sumx * sumy)
    divisor =  n*sumx2-sumx**2
    b = dividend/divisor
    meanx = sumx/n
    meany = sumy/n
    a = meany - b * meanx
    curA = 1 / a
    curB = b * curA
    c_dividend = (n * sumxy) - (sumx * sumy)
    c_divisor = math.sqrt(((n* sumx2) - sumx**2) * ((n * sumy2) - sumy**2))
    r= c_dividend / c_divisor
    print(f"y = ({round(curA, 4)}) (x /(({round(curB, 4)}) + x)) ")
    print(f"Hyperbolic Regression is {round(r, 4)}")


sl = data['sepal.length']
sw = data['sepal.width']
pl = data['petal.length']
pw = data['petal.width']
n = len(data)
#pair 1 Sepal Length VS Sepal Width 
print("Sepal Length VS Sepal Width")
linear_regre(sl,sw,n)
exp_regre(sl,sw,n)
log_regre(sl,sw,n)
hyp_regre(sl,sw,n)

print("")
#pair 2 Sepal Width VS Petal Width 
print("Sepal Width VS Petal Width")
linear_regre(sw,pw,n)
exp_regre(sw,pw,n)
log_regre(sw,pw,n)
hyp_regre(sw,pw,n)

print("")
#pair 3 Sepal Length VS Petal Length 
print("Sepal Length VS Petal Length")
linear_regre(sl,pl,n)
exp_regre(sl,pl,n)
log_regre(sl,pl,n)
hyp_regre(sl,pl,n)

print("")
#pair 4 Petal Length VS Petal Width 
print("Petal Length VS Petal Width")
linear_regre(pl,pw,n)
exp_regre(pl,pw,n)
log_regre(pl,pw,n)
hyp_regre(pl,pw,n)

print("")
#pair 5 Petal Length VS Sepal Width 
print("Petal Length VS Sepal Width")
linear_regre(pl,sw,n)
exp_regre(pl,sw,n)
log_regre(pl,sw,n)
hyp_regre(pl,sw,n)

print("")
#pair 6 Petal Width VS Sepal Length 
print("Petal Width VS Sepal Length")
linear_regre(pw,sl,n)
exp_regre(pw,sl,n)
log_regre(pw,sl,n)
hyp_regre(pw,sl,n)   
```
