```python
import math
import cmath
import pandas as pd
data = pd.read_csv('iris.csv')
def correff_r(raw_x, raw_y, raw_count):
    x_sum = 0
    y_sum = 0
    xy_sum = 0
    xsq_sum = 0
    ysq_sum = 0
    dividend_correff = 0
    divisor_correff = 0
    r_value = 0
    # getting sum of x
    for ints_x in raw_x:
        x_sum += float(ints_x)
    # getting sum of y
    for ints_y in raw_y:
        y_sum += float(ints_y)
    # getting sum of x*y
    for ints_sum in range(int(raw_count)):
        xy_sum += (raw_x[ints_sum] * raw_y[ints_sum])
    # getting sum of x squared
    for powx_sum in raw_x:
        xsq_sum += int(powx_sum**2)
    # getting sum of y squared
    for powy_sum in raw_y:
        ysq_sum += int(powy_sum**2)
    # dividend of correlation coefficient
    dividend_correff += ((int(raw_count) * xy_sum) - (x_sum * y_sum))
    # divisor of correlation coefficient
    divisor_correff += cmath.sqrt(((int(raw_count) * xsq_sum) - x_sum**2) * ((int(raw_count) * ysq_sum) - y_sum**2))
    # getting r
    r_value += dividend_correff / divisor_correff
    print(f"Correlation Coefficient Regression Value: {r_value}")


def linear_regression(raw_linear_x, raw_linear_y, raw_count):
    x_sum = 0
    y_sum = 0
    sum_xy = 0
    xsq_sum = 0
    sum_xsq = 0
    dividend_regre = 0
    divisor_regre = 0
    formula_a_regre = 0
    formula_b_regre = 0
    # getting sum of x
    for ints_x in raw_linear_x:
        x_sum += float(ints_x)
    # getting sum of y
    for ints_y in raw_linear_y:
        y_sum += float(ints_y)
    # getting sum of x*y
    for ints_sum in range(int(raw_count)):
        sum_xy += (raw_linear_x[ints_sum] * raw_linear_y[ints_sum])
    # getting sum of x squared
    for pow_sum in raw_linear_x:
        xsq_sum += (pow(pow_sum, 2))
    # getting mean of x
    mean_x = x_sum / int(raw_count)
    # getting mean of y
    mean_y = y_sum / int(raw_count)
    # getting sum of x then squaring
    sum_xsq += x_sum**2
    # dividend of regression formula
    dividend_regre += ((int(raw_count) * sum_xy) - (x_sum * y_sum))
    # divisor of regression formula
    divisor_regre += ((int(raw_count) * xsq_sum) - sum_xsq)
    # formula B in linear regression
    formula_b_regre += dividend_regre / divisor_regre
    # formula A in linear regression
    formula_a_regre += (mean_y - (formula_b_regre * mean_x))
    # outputs
    print(f"Mean of X: {mean_x}")
    print(f"Mean of Y: {mean_y}")
    print(f"Sum of X*Y: {sum_xy}")
    print(f"Sum of X2: {xsq_sum}")
    print(f"Linear Regression: y = {formula_a_regre} + {formula_b_regre}x")


def exponential_regression(raw_exp_x, raw_exp_y, raw_count):
    x_sum = 0
    y_sum = 0
    lny_sum = 0
    mean_x = 0
    mean_y = 0
    mean_lny = 0
    sum_xsq = 0
    sum_xlny = 0
    dividend_regre = 0
    divisor_regre = 0
    formula_a_regre = 0
    formula_b_regre = 0
    # getting sum of x
    for ints_x in raw_exp_x:
        x_sum += float(ints_x)
    # getting sum of y
    for ints_y in raw_exp_y:
        y_sum += float(ints_y)
    # getting mean of x
    mean_x += x_sum / int(raw_count)
    # getting mean of y
    mean_y += y_sum / int(raw_count)
    # natural logarithm sum of y
    for ln_ints_y in raw_exp_y:
        lny_sum += math.log(float(ln_ints_y))
    # getting mean of lny
    mean_lny += lny_sum / int(raw_count)
    # getting square of x sum
    for pow_sum in raw_exp_x:
        sum_xsq += (pow(pow_sum, 2))
    # getting sum of X*lnY
    for ints_sum in range(int(raw_count)):
        sum_xlny += (raw_exp_x[ints_sum] * math.log(raw_exp_y[ints_sum]))
    # dividend of regression formula
    dividend_regre += ((int(raw_count) * sum_xlny) - (x_sum * lny_sum))
    # divisor of regression formula
    divisor_regre += ((int(raw_count) * sum_xsq) - pow(x_sum, 2))
    # formula B in regression
    formula_b_regre += dividend_regre / divisor_regre
    print(mean_x)
    print(mean_y)
    print(mean_lny)
    print(sum_xsq)
    print(sum_xlny)


def logarithmic_regression(raw_log_x, raw_log_y, raw_count):
    logx_sum = 0
    logy_sum = 0
    logx_mean = 0
    logy_mean = 0
    logxsq_sum = 0
    logysq_sum = 0
    sum_logxlogy = 0
    dividend_regre = 0
    divisor_regre = 0
    formula_a_regre = 0
    formula_b_regre = 0
    # getting log x sum
    for ints_x in raw_log_x:
        logx_sum += math.log(float(ints_x), 10) * 10
    # getting log y sum
    for ints_y in raw_log_y:
        logy_sum += math.log(float(ints_y), 10) * 10
    # getting log x mean
    logx_mean += logx_sum / int(raw_count)
    # getting log y mean
    logy_mean += logy_sum / int(raw_count)
    # getting sum of x squared
    for powx_sum in raw_log_x:
        logxsq_sum += math.log(float(powx_sum), 10)**2
    # getting sum of y squared
    for powy_sum in raw_log_y:
        logysq_sum += math.log(float(powy_sum), 10)**2
    # getting sum of logx * logy
    for sum_logs in range(int(raw_count)):
        sum_logxlogy += ((math.log(float(raw_log_x[sum_logs]), 10)) * (math.log(float(raw_log_y[sum_logs]), 10)))
    # dividend of regression formula
    dividend_regre += ((int(raw_count) * sum_logxlogy) - (logx_sum * logy_sum))
    # divisor of regression formula
    divisor_regre += ((int(raw_count) * logxsq_sum) - pow(logx_sum, 2))
    # formula B in regression
    formula_b_regre += dividend_regre / divisor_regre
    print(logx_mean)
    print(logy_mean)
    print(logxsq_sum)
    print(logysq_sum)
    print(sum_logxlogy)


def hyperbolic_regression(raw_hyp_x, raw_hyp_y, raw_count):
    hypx_sum = 0
    hypy_sum = 0
    hypx_mean = 0
    hypy_mean = 0
    hypx_sq_sum = 0
    hypy_sq_sum = 0
    sum_hypxhypy = 0

    # getting 1/x
    for ints_x in raw_hyp_x:
        hypx_sum += 1 / float(ints_x) * 10
    # getting 1/y
    for ints_y in raw_hyp_y:
        hypy_sum += 1 / float(ints_y) * 10
    # getting mean of 1/x
    hypx_mean += hypx_sum / int(raw_count)
    # getting mean of 1/y
    hypy_mean += hypy_sum / int(raw_count)
    for powhyp_x in raw_hyp_x:
        hypx_sq_sum += (1 / float(powhyp_x))**2
    for powhyp_y in raw_hyp_y:
        hypy_sq_sum += (1 / float(powhyp_y))**2
    for sum_hyps in range(int(raw_count)):
        sum_hypxhypy += ((1 / float(raw_hyp_x[sum_hyps])) * (1 / float(raw_hyp_y[sum_hyps])))
    print(hypx_mean)
    print(hypy_mean)
    print(hypx_sq_sum)
    print(hypy_sq_sum)
    print(sum_hypxhypy)


slength = data['sepal.length']
swidth = data['sepal.width']
plength = data['petal.length']
pwidth = data['petal.width']
n=len(data)
#pair 1 Sepal Length VS Sepal Width 
linear_regression(slength,swidth,n)
exponential_regression(slength,swidth,n)
logarithmic_regression(slength,swidth,n)
hyperbolic_regression(slength,swidth,n)


```
