```python

from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')


iris = load_iris()
X = iris.data
y = iris.target
iris_df = pd.DataFrame(iris['data'], columns=iris['feature_names'])
iris_df['species'] = iris['target']

knn = KNeighborsClassifier(n_neighbors=6)
knn.fit(X, y)
print("Enter the characteristics of your iris")
a = float(input("sepal length:"))
b = float(input("sepal Width:"))
c = float(input("petal length:"))
d = float(input("petal width:"))
value = knn.predict([[a, b, c, d]])
if (value == 0):
    species = "Setosa"
elif (value == 1):
    species = "Versicolor"
elif (value == 2):
    species = "Virginia"


def scatterplot(xlabel, ylabel, valx, valy, sp, n):
    colours = ['gray', 'orange', 'blue'] 
    species = ['Setosa', 'Versicolor', 'Virginica']
    plt.figure(n, figsize=(8, 5))
    plt.clf()
    for i in range(0, 3):
        species_df = iris_df[iris_df['species'] == i]
        plt.scatter(
            species_df[xlabel],
            species_df[ylabel],
            color=colours[i],
            label=species[i],
            edgecolor='k'

        )                   
    plt.scatter(valx, valy, color="red", edgecolor='k', label=xlabel + " = " +
                str(valx) + ", " + ylabel + " = " + str(valy) + ", output = " + sp)
    plt.title(xlabel + " vs " + ylabel)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend(loc=4, prop={'size': 8})


sl = "sepal length (cm)"  # a
sw = "sepal width (cm)"  # b
pl = "petal length (cm)"  # c
pw = "petal width (cm)"  # d

scatterplot(sl, sw, a, b, species, 1)
scatterplot(sl, pl, a, c, species, 2)
scatterplot(sl, pw, a, d, species, 3)
scatterplot(sw, pl, b, c, species, 4)
scatterplot(sw, pw, b, d, species, 5)
scatterplot(pl, pw, c, d, species, 6)
plt.show()
```
